## Importation des données

```python
import pandas as pd
import s3fs

BUCKET = "xifar19"
FILE_KEY_S3 = "boardgames/bgg_dataset.csv"
FILE_PATH_S3 = f"{BUCKET}/{FILE_KEY_S3}"

fs = s3fs.S3FileSystem()

with fs.open(FILE_PATH_S3, mode="rb") as file_in:
    df = pd.read_csv(file_in, sep=";")
```
